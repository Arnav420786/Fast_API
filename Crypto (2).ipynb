```python
service = Service(executable_path=path)
```

```python
driver = webdriver.Chrome(service=service)
```

```python
website = 'https://www.coingecko.com/'
```

```python
driver.get(website)#it loads the website's window 
```

```python
#driver.quit()#it quits 
```

```python
tag_names = driver.find_elements(By.TAG_NAME, "tr")
data =[]
for tag in tag_names:
    cols = tag.find_elements(By.TAG_NAME, "td")  
    cols_text = []
    for col in cols:
        cols_text.append(col.text)
    data.append(cols_text)
print(data)
```

```python
df = pd.DataFrame(data)
```
